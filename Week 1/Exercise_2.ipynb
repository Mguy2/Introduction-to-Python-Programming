## 2. Prime Search

*Prime* numbers are integers that can only be divided by 1 and itself.<br>
For instance, 2, 3, and 5 are *prime* numbers, but 4, 6, and 10 are not.
A simple, yet not effective, way to discover works like this:
* Select a candidate number
* Check if such number is divisible between another number smaller than itself
* If there is no such divider, the number is prime

For this exercise create a function `prime_search` that uses two input arguments, `i` and `end` (both integer variables) denoting the start and end number. For instance, you could use this function to print out all prime numbers between `i = 2` amd `end = 100`.

**Note:** Your code should work with any two numbers.

**Hint:** consider using nested `while`-loops and `break` to solve this problem.

**Example:**  
*Input:*  
```python
i = 2
end = 100
```

*Output:*  
<pre>
2 is prime
3 is prime
5 is prime
7 is prime
    .
    .
    .
97 is prime
</pre>

````python
#// BEGIN_TODO [primes] Prime Search 

def prime_search(i :int, end :int) -> None:
    """random docstring"""
    while i <= end:
        prime_test(i)
        i += 1

def prime_test(input: int) -> None:
    """Random docstring."""
    i:int = input - 1
    if input <= 0:
        return
    while i > 1:
        check = input % i
        if check == 0:
            return
        else:
            check = 0
            i -= 1
    print(input,"is prime")

prime_search(10, 100)
#// END_TODO [primes]
````

## 3. Collatz Problem

We can use `while`-loops to test mathematical problems such as the *Collatz* problem.<br>
The *Collatz* problem is an easy iterated formula in which you have an *integer* variable called `integer` and has the following formula:
* If the `integer` is even: Divide the integer by 2
* If the `integer` is odd: Multiply the integer by 3 and add 1
* Do this until the sequence reaches 1 (After which it will loop 1, 4, 2, 1, ...)

For this exercise, create a function called `collatz` that uses an *int* input argument called `integer` and prints out the sequence of the collatz problem (see the example output below).

**Note:** The question of the *Collatz* problem is whether every starting value will eventually reach 1. 
This has not been formally proven yet, but it is assumed to be true.

**Hint:** You need to create a program using a `while`-loop that prints the *Collatz* sequence of an *integer*.



**Example:**  
*Input:*  
```python
integer = 20
```

*Output:*  
<pre>
10
5
16
8
4
2
1
</pre>

````python
#// BEGIN_TODO [Collatz] Collatz Problem 

def iseven(integer :int) -> int:
    """Random docstring."""
    return integer % 2 == 0

def collatz(integer :int) -> None:
    """Random docstring."""
    while integer != 1:
        output = iseven(integer)
        if output == True:
            integer = integer // 2
        if output == False:
            integer = integer * 3 + 1
        print(integer)

#// END_TODO [Collatz]
````

## 5. Writing Books

We would like to print all chapters and sections for a book.<br>

For this exercise create a function called `print_chapters`, which has `start_chapter`, `final_chapter`, and `max_section` as *integer* parameters:

* `start_chapter` is the first chapter you are going to print.
* `final_chapter` is the final chapter you are going to print.
* `max_section` is the number of sections each chapter has - All chapters have the same number of `max-section`.

**Hint:** `print_chapters` has to print all chapters and sections within each chapter.

**Example:**  
*Input:*  
```python
start_chapter = 1
final_chapter = 3
max_section = 2
```

*Output:*  
<pre>
Chapter 1
Section 1.1 
Section 1.2 
Chapter 2 
Section 2.1 
Section 2.2 
Chapter 3 
Section 3.1 
Section 3.2 
</pre>

````python
#// BEGIN_TODO [books] Writing Books

def print_chapters(start_chapter :int, final_chapter :int, max_section :int) -> None:
    """Random docstring."""
    chap :int = 1
    sect :int = 1
    section_text = "Section {chapter}.{section}"
    if start_chapter > 0 and final_chapter > 0 and max_section >= 0 and start_chapter < final_chapter:
        while chap <= final_chapter:
            print("Chapter", chap)
            while sect <= max_section:
                print(section_text.format(chapter = chap, section = sect))
                sect += 1
            sect = 1
            chap += 1
    else:
        print("Invalid Input")

#// END_TODO [books]
````

## 6. Insurance Calculations

We would like to check what price someones car insurance will be.<br>
This is based on the total amount of light and heavy accidents that the person made over the last year, and the number of years of driving experience.

For this exercise, we need to create a function that checks the price as often as we want, because we need to check it for multiple people. 

We have already made a skeleton of the function called `checkPrice` (so there is no code within the function), your job is to finish the function such that:
* The base insurance is 2,000 a year
* Every light accident increases the price by 100
* Every heavy accident increases the total price by 10%

For instance: If the insurance premium was 100 (before any heavy accidents), and one has the first heavy accident, the premium increases by $10\%, i.e. 100 * 1.10 = 110$. The second heavy accident increases the premium then by another $10\%$. Thus, $110 * 1.10 = 121$.

* Every year of driving experience drops the price by 200, but only after the first 5 learner years.
* The price of the insurance can never be lower than 800

**Hint:** `checkPrice` has to return the price of the insurance.

**Example:**  
*Input:*  
```python
light_accidents = 0
heavy_accidents = 0
driving_experience = 0
```

*Output:*  
<pre>
2000.0
</pre>

````python
#// BEGIN_TODO [Insurance] Insurance Calculations

def checkPrice(light_accidents :int, heavy_accidents :int, driving_experience :int) -> float:
    """Random docstring."""
    base = 2000
    acci_small = 100
    acci_big = 1.1
    discount = 200
    output = base
    if light_accidents >= 68373396 or heavy_accidents >= 68373396 or driving_experience >= 130:
        return
    if driving_experience >= 5: 
        output -= ((driving_experience - 5) * discount)
    if light_accidents > 0:
        output += (acci_small * light_accidents)
    if heavy_accidents > 0:
        output *= (heavy_accidents * acci_big)
    if output < 800:
        return 800 / 1
    else:
        return output / 1

checkPrice(0, 0, 0)


#// END_TODO [Insurance]
````

## 7. Sorting

For this exercise create a function called `sorting`, which takes three *integers* as arguments.<br>
This function should sort in ascending order and return the numbers in a string. Numbers should be separated by *commas*, and one *space* is required after each *comma*.

**Note:** 
- This function returns something and does not print.

**Hint:**
- Use `max` and `min` functions instead of using `if`-statements to solve this problem.

**Example:**  
*Input:*  
```python
num1 = 4
num2 = 7
num3 = 2
```

*Output:*  
<pre>
2, 4, 7
</pre>

````python
#// BEGIN_TODO [sorting_numbers] Sorting 

def sorting(num_1 :int, num_2 :int, num_3 :int) -> str:
    """Random docstring."""
    min_int = min(num_1, num_2, num_3)
    max_int = max(num_1, num_2, num_3)
    mid_int = (num_1 + num_2 + num_3) - (min_int + max_int)
    sorted = "{min_int}, {mid_int}, {max_int}"
    return sorted.format(min_int = min_int, mid_int = mid_int, max_int = max_int)

#// END_TODO [sorting_numbers]
````

## 8. Multi Multiplication

Functions can have all possible variable types as parameters.<br>

* The second function called is `run_triple`. It will take the function  `multiply` as an argument and besides this, four other arguments, `num1`, `num2`, `num3` and `num4`. `run_triple` calls the functions `multiply` twice, once with the values `num1` and `num2` and once with the arguments `num3` and `num4`. ## 9. String Cleaning

Your database is storing the names of customers.<br>
However, because of the way we use these names, the stored names have to follow certain rules. 

For this exercise create a function called `standardize_name` that changes the name if it does not comply to the rules, into names that do. 
Return the changed name, or the original name when it does follow the naming rules.

The rules for the names are:
* Start with capital letter
* All other letters are lowercase
* Only the characters a-z A-Z are allowed in the name, any other letters have to be deleted
* Spaces and tabs are not allowed, underscores are used instead
* Names can be no longer than 80 characters, longer names will lose their extra letters

**Example:**  
*Input:*  
```python
name = "python3 cOUrse"
```

*Output:*  
<pre>
Python_course
</pre>

````python
#// BEGIN_TODO [transforming_names] String Cleaning 

def standardize_name(name :str) -> str:
    """Random docstring."""
    i :int = 0
    j :int = 0
    rm_str :int = 0
    str_len :int = len(name)   
    name = name.replace(' ', '_')
    name = name.replace('   ', '_')
    while j < str_len:
        if int(ord(name[j])) > 64 and int(ord(name[j])) < 91:
            name = name.replace(chr(j), chr(int(ord(name[j])) + 32))
        if (int(ord(name[j])) < 65 or (int(ord(name[j])) > 90 and int(ord(name[j])) < 97) or int(ord(name[j])) > 122) and not int(ord(name[j]) == 95):
            rm_str = len(name)
            name = name.replace(name[j], '')
            str_len = (str_len - (rm_str - len(name)))
            j -= 1
        j += 1
    while ord(name[i]) == 95:
        i += 1
    name = name[i:(i+80)]
    name = name.capitalize()
    return(name)

standardize_name("   ")
    
#// END_TODO [transforming_names]
````

## 10. Counting Letters

For this exercise create a function called `count_case` that accepts a *string* and returns the number of upper and lower case letters as a string. 

**Example:**  
*Input:*  
```python
"I am currently attending Programing Course"
```

*Output:*  
<pre>
Upper case: 3 - Lower case: 34
</pre>

````python
#// BEGIN_TODO [count_case] Counting Letters 

def count_case(string :str) -> str:
    """Random docstring."""
    i = len(string) - 1
    big :int = 0
    small :int = 0
    text = "Upper case: {big} - Lower case: {small}"
    while i >= 0:
        if int(ord(string[i])) > 64 and int(ord(string[i])) < 91:
            big += 1
        if int(ord(string[i])) > 96 and int(ord(string[i])) < 123:
            small += 1
        i -= 1
    output = text.format(big = big, small = small)
    return output

#// END_TODO [count_case]
````
